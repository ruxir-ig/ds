```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# load inbuilt dataset
titanic = pd.read_csv('/content/titanic.csv')

# view first rows
print(titanic.head())

```

```python
# ---------------------------
# Pattern 1: Survival by Gender
# ---------------------------
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival Count by Gender")
plt.show()

```

```python
# ---------------------------
# Pattern 2: Survival by Passenger Class
# ---------------------------
sns.countplot(x='pclass', hue='survived', data=titanic) # Changed 'class' to 'pclass'
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# ---------------------------
# Pattern 3: Age Distribution
# ---------------------------
sns.histplot(titanic['age'].dropna(), bins=30)
plt.title("Age Distribution")
plt.show()

```

```python
# ---------------------------
# Histogram for Fare Distribution
# ---------------------------
sns.histplot(titanic['fare'], bins=30)
plt.title("Fare Distribution")
plt.xlabel("Ticket Fare")
plt.ylabel("Passenger Count")
plt.show()
```
